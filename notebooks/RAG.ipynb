```python
from datasets import load_dataset

dataset = load_dataset("csv", data_files={"train": "/content/Spotify 2010 - 2019 Top 100.csv"})

print(dataset["train"].features)
```

```python
import pandas as pd

# Load dataset
df = dataset["train"].to_pandas()
print(df[['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val']].describe())

```

```python

# Select relevant columns
df = df.dropna()
# Define mood categories
mood_map = {
    'anger': 'intense', 'boredom': 'chill', 'empty': 'melancholic', 'enthusiasm': 'energetic',
    'fun': 'playful', 'happiness': 'happy', 'hate': 'dark', 'love': 'romantic',
    'neutral': 'neutral', 'relief': 'relaxing', 'sadness': 'sad', 'surprise': 'unexpected',
    'worry': 'anxious'
}

# Assign mood based on valence
def classify_mood(row):
    val = row['val'] / 100  # Normalize to 0-1 range
    energy = row['nrgy'] / 100  # Normalize
    danceability = row['dnce'] / 100  # Normalize
    loudness = row['dB']  # No need to normalize

    if val > 0.75 and energy > 0.7 and danceability > 0.7:
        return "happiness"
    elif val < 0.35 and energy < 0.4:
        return "sadness"
    elif energy > 0.85 and loudness > -5:
        return "anger"
    elif danceability > 0.75 and energy > 0.65:
        return "fun"
    elif val > 0.6 and energy < 0.5:
        return "love"
    elif loudness < -10 and val < 0.3:
        return "empty"
    elif energy < 0.4 and val > 0.5:
        return "relief"
    elif danceability < 0.4 and energy < 0.3:
        return "boredom"
    elif val > 0.5 and danceability > 0.6 and energy < 0.5:
        return "enthusiasm"
    elif val < 0.4 and energy > 0.7 and loudness > -6:
        return "hate"
    elif val < 0.5 and energy > 0.6 and loudness > -6:
        return "surprise"
    elif energy < 0.6 and val < 0.6:
        return "neutral"
    else:
        return "worry"

# Apply to dataset
df['mood'] = df.apply(classify_mood, axis=1)
df['text'] = df.apply(lambda x: f"You should listen to {x['title']} by {x['artist']}. Genre: {x['top genre']}. Mood: {x['mood']}.", axis=1)

df.head()
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Load embedding model
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Convert song texts into embeddings
song_embeddings = embedding_model.encode(df['text'].tolist(), convert_to_tensor=True)

# Convert embeddings to numpy array
song_embeddings_np = np.array([embedding.cpu().numpy() for embedding in song_embeddings])

# Initialize FAISS index
index = faiss.IndexFlatL2(song_embeddings_np.shape[1])
index.add(song_embeddings_np)
```

```python
import random
def retrieve_song(mood_keyword, top_k=5):
    # Map input mood to a related mood category
    mapped_mood = mood_map.get(mood_keyword, 'neutral')

    # Encode the mapped mood
    mood_embedding = embedding_model.encode([mapped_mood], convert_to_tensor=True)
    mood_embedding_np = np.array(mood_embedding.cpu().numpy())

    # Retrieve most relevant songs
    _, indices = index.search(mood_embedding_np, top_k)
    recommended_songs = df.iloc[indices[0]][['text']].values.tolist()
    selected_song = random.choice(recommended_songs)

    return selected_song[0]
```

```python
print(retrieve_song("happiness"))
print(retrieve_song("worry"))
```
